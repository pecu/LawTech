```python
import plotly
import plotly.graph_objs as go

import pandas as pd
```

```python
data = pd.read_csv("LawTech_0727_Pandas.csv",index_col=0)
data.head()
```

```python
location = []
year = []
for i in data["title"]:
    a = i.split(" ")
    location.append(a[0])
    year.append(a[1])
```

```python
data["year"] = year
data["location"] = location
```

```python
data.head()
```

```python
j = list(set(data["judge"]))
print(j)
```

```python
judge = list(data["judge"])

for i in j:
    print(judge.count(i),end=" ")
    
judge_count= [] 
for i in j:
    judge_count.append(judge.count(i))
```

```python
plotly.offline.init_notebook_mode(connected=True)

plotly.offline.iplot({
    "data": [go.Scatter(x=j, y=judge_count)],
    "layout": go.Layout(title="判決數量視覺化")
})
```
