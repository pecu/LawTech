<h1>Step3 : 資料整理</h1>

```python
a = data["法規"][0]
print(a)
print(type(a))
```

```python
a = eval(data["法規"][0])
print(a)
print(type(a))
```

```python
laws = []
for i in data["法規"]:
    laws.append(eval(i))
data["法規"] = laws
data.head()
```

```python
# ref : https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html
# ref : https://pandasguide.readthedocs.io/en/latest/Pandas/overview.html
```

```python
import re 
new_content = []
for i in data["內文"]:
    a = re.sub("\n","",i)
    a = re.sub("\u3000","",a)
    a = re.sub(" ","",a)
    a = re.sub("\uf67e","",a)
    a = re.sub("\uf67d","",a)
    a = re.sub("\uf67c","",a)
    a = re.sub("\uf679","",a)
    new_content.append(a)
data["內文"] = new_content
data["內文"][0]
```

```python
date = []
for i in range(len(data)):
    a = data["內文"][i]
    time = a.split("國")[-1]
    time = time.split("日")[0] + "日"
    date.append(time)

data["判決時間"] = date
data.head(3)
```

<h1>Step4 : 格式套用</h1>

```python
def get_date(data):
    time = data["內文"].split("中華民國")[-1].split("日")[0] + "日"
    return time

a = data[:5].apply(get_date,axis=1)
print(a)
```

```python
def get_date2(data):
    time = data.split("中華民國")[-1].split("日")[0] + "日"
    return time

a = data[:5].apply(lambda a:get_date2(a["內文"]),axis=1)
print(a)
```
