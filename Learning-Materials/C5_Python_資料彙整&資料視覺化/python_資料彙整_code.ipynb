```python
a = eval(data["法規"][0])
print(a)
print(type(a))
```

```python
laws = []
for i in data["法規"]:
    laws.append(eval(i))
data["法規"] = laws
data.head()
```

```python
import re 
new_content = []
for i in data["內文"]:
    a = re.sub("\n","",i)
    a = re.sub("\u3000","",a)
    a = re.sub(" ","",a)
    a = re.sub("\uf67e","",a)
    a = re.sub("\uf67d","",a)
    a = re.sub("\uf67c","",a)
    new_content.append(a)
data["內文"] = new_content
data["內文"][0]
```

```python
w = '法院得逕以裁定駁回上訴。中華民國105年11月24日書記官連歆喬'
```

```python
date = []
for i in range(len(data)):
    a = data["內文"][i]
    a1 = re.findall(r"中華民國\d+年\d+月\d+日",a)
    a2 = a1[0].split("國")[1]
    date.append(a2)

data["判決時間"] = date
data.head(5)
```

## **Step4 : 格式套用**

```python
def get_date(data):
    a1 = re.findall(r"中華民國\d+年\d+月\d+日",data["內文"])
    a2 = a1[0].split("國")[1]
    return a2

a = data[:5].apply(get_date,axis=1)
print(a)
```

```python
def get_date2(data):
    a1 = re.findall(r"中華民國\d+年\d+月\d+日",data)
    a2 = a1[0].split("國")[1]
    return a2

a = data[:5].apply(lambda a:get_date2(a["內文"]),axis=1)
print(a)
```
