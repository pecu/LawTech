```python
# 2. 轉換案號成新的3個欄位，並加入到資料中(簡易)
year = []
locate_type = []
number = []

for i in data["案號"]:
    item = i.split(",")
    year.append(item[0])
    locate_type.append(item[1])
    number.append(item[2])
    
data["案號(年)"] = year
data["案號(字)"] = locate_type
data["案號(號)"] = number
data.head()
```

```python
# 3. 計算每個裁判共使用的法規數量，並新增一欄加入資料(簡易)
laws = []
for i in data["法規"]:
    laws.append(eval(i))
data["法規"] = laws

law_number = []
for i in data["法規"]:
    num = len(i)
    law_number.append(num)
data["使用法規數量"] = law_number
data.head()
```

```python
# 4. 使用apply、lamdba方式，自訂一個規則，並套用到所有資料中(中等)
from datetime import datetime
import re

def change_time(d):                                                           #d填入時間，格式是XX年XX月XX日 
    match = re.search("(?P<year>\d+)年(?P<month>\d+)月(?P<date>\d+)日",d["判決時間"])     #利用re套件取出日期數字部分
    year = int(match.group("year"))+1911                                      #處理年分+1911轉換成西元
    date = str(year) + "-" +match.group('month') + "-" + match.group('date')  #將時間拼起來成新的字串
    date = datetime.strptime(date, "%Y-%m-%d")
    return date

a = data[:5].apply(change_time,axis=1)
print(a)
```

```python
from datetime import datetime
import re
def change_time(d):                                                           #d填入時間，格式是XX年XX月XX日 
    match = re.search("(?P<year>\d+)年(?P<month>\d+)月(?P<date>\d+)日",d)     #利用re套件取出日期數字部分
    year = int(match.group("year"))+1911                                      #處理年分+1911轉換成西元
    date = str(year) + "-" +match.group('month') + "-" + match.group('date')  #將時間拼起來成新的字串
    date = datetime.strptime(date, "%Y-%m-%d")
    return date

a = data[:5].apply(lambda a:change_time(a["判決時間"]),axis=1)
print(a)
```
