## Tuple
---
Tuple是利用小括號（ ）將資料包起來，為「有順序性」的「不可變動」列表，可放入不同型態之資料

### Tuple的建立

```python
#方法一：直接指派資料，並且以逗號區隔，Python會將此預設為Tuples。
print("方法一")
T1 = 1,2,3
print(T1)
T2 = "Hello", "Tuple", 3
print(T2)

##如果只有一個資料的話要加逗號（,），不然會被當作int存取
T3 = 3,
print(T3)

#方法二：於 () 符號中輸入資料，並且以逗號區隔。
print("方法二")
T4 = (4,5,6)
print(T4)
T5 = ("Hello", "Tuple", 3)
print(T5)

#方法三：利用tuple()將List型態轉換成Tuple
print("方法三")
L1 = ["This","is","a","List"]
print(L1, type(L1))
T5 = tuple(L1)
print(T5, type(T5))
```

### Tuple元素的取得

```python
#利用index(索引)取得單個元素
names_1 = ("Tom", "Dan", "Amy", "Sam", "Tom")
print(names_1)
print(names_1[0])

#取得特定範圍
print(names_1[1:3])
print(names_1[2:-1])
print(names_1[:4])
print(names_1[2:])
```

### Tuple的其他特性

```python
#合併兩個Tuples
names_2 = ("小明","小王")
group = names_1 + names_2
print(group)

#看資料長度 -> len()
print("group中有多少人：", len(group))

#利用index(取得該元素的位置)
print("Dan在第幾個位置：", group.index("Dan"))

#查看該元素是否在Tuple中 -> in、not in
print("Anna在group中：", "Anna" in group)
print("Anna不在group中：", "Anna" not in group)
```

## List
---
List是利用中括號 [ ] 將資料包起來，為「有順序性」的「可變動」列表，可放入不同型態之資料

### List的建立

```python
#方法一：將資料放入中括號內指派
L1 = [1, 2, "3", "四"]
print(L1)

#方法二：利用list()轉換型態
group_list = list(group)
print(group, type(group))
print(group_list, type(group_list))
```

### List元素的取得

```python
#利用index(索引)取得單個元素
print(group_list)
print(group_list[2])
print(group_list[-2])

#取得特定範圍
print(group_list[1:3])
print(group_list[2:-1])
print(group_list[:4])
print(group_list[2:])
```

### List的新增、修改、刪除

#### 「新增」

```python
subjects = ["國文", "英文", "社會"]
print(subjects)

#方法一：利用append()將元素新增至最後一個位置
subjects.append("自然")
print(subjects)

#方法二：利用insert()將元素新增至指定位置
subjects.insert(2, "數學")
print(subjects)
```

#### 「修改」

```python
subjects[4] = ["化學", "物理"]
print(subjects)
```

#### 「刪除」

```python
print(subjects)
#方法一：利用pop()將最後一個元素刪除，或是傳入想刪除的位置
subjects.pop()
print(subjects)
subjects.pop(2)
print(subjects)

#方法二：利用del刪除一段範圍的元素
L1 = [1,2,3,3,4,5,6,7,8,9,10]
print(L1)
del L1[4:-2]
print(L1)

#方法三：利用remove()刪除指定元素，若傳入的有多個重複，只會刪除第一個出現的
L1.remove(3)
print(L1)
```

### List的其他特性

```python
#合併兩個List
grade_1 = [90, 95, 55]
grade_2 = [88, 60, 55]
grade = grade_1 + grade_2
print(grade)

#看資料長度 -> len()
print("有幾科成績：", len(grade))

#利用index(取得該元素的位置)
print("88分在第幾個位置：", grade.index(88))

#查看該元素是否在List中 -> in、not in
print("有100分的成績：", 100 in group)
print("沒有100分的成績：", 100 not in group)
```

## Set
---
以大括號{ }表示，為「無順序性」的「無重複」集合，可放入不同型態之資料，但每項資料必須是固定的內容

### Set的建立

```python
s1 = {1, 2, 3, "set"}
print(s1, type(s1))

#包含tuple的set
##因為set沒有順序的概念，印出來的資料順序不固定
s2 = {5, "8", ("a", "b", "c")}
print(s2, type(s2)) 

#set中不可放入list資料，會報錯
#s3 = {5, "8", ["a", "b", "c"]}  
```

### Set元素的取得
因為set沒有順序的概念，不像List、Tuple能用index的方法取得

### Set的新增、刪除

#### 「新增」

```python
#方法一：利用add()新增單筆資料
print(s2)
s2.add(1000)
print(s2)

#方法二：利用update()一次新增多筆筆資料
s_new = {"New", "Set"}
print(s_new)
s2.update(s_new)
print(s2)
```

#### 「刪除」

```python
#方法一：利用pop()將最後一個元素刪除
s2.pop()
print(s2)

#方法二：利用remove()刪除指定元素
s2.remove(('a', 'b', 'c'))
print(s2)
```

### Set的運算
利用集合運算取得兩集合的交集（&）、聯集（|）、差集（-）、反交集（^）

```python
s3 = {1, 2, 3, 4}
s4 = {3, 4, 5, 6}

#取交集
print(s3 & s4)

#取聯集
print(s3 | s4)

#取差集，順序有差
print(s3 - s4)
print(s4 - s3)

#取反交集
print(s3 ^ s4)
```

### Set的其他特性

```python
#用set()將字串轉變為集合
s_str = set("animals")
print(s_str)

#用len()取得長度
print(len(s_str))

#查看資料是否在集合中
print("m" in s_str)
print("m" not in s_str)
```

## Dict
---
Dictionary是利用鍵值對（Key-value Pair）的方式存取資料，以大括號{ }表示，為「無順序性」的「可變動」容器，可放入不同型態之資料

### Dict的建立

```python
#方法一：於 {} 符號中輸入每個元素的鍵(Key)與值(Value)
animals = {1 : "狗", 2 : "貓", 3 : "羊"}
print(animals, type(animals))

#方法二：使用dict()方法，傳入鍵(Key)的名稱，並且指派值(Value)給它
heights = dict(Mike = 185, Sam = 178, Amy = 163) 
print(heights, type(animals))
```

### Dict元素的取得

```python
#利用key去查找value
print(animals[1])
print(heights["Mike"])

#沒辦法透過index方式查找，會報錯
#print(heights[0])

#取得字典中的字典
new_dict = dict(Mike = 185, Sam = 178, Amy = 163, animal = animals) 
print(new_dict)
print(new_dict['animal'])
print(new_dict['animal'][2])
```

### Dict的新增、修改、刪除

#### 「新增」

```python
heights['Peter'] = 166
print(heights)

#方法二：利用update()將多筆資料更新至原dict中
heights_2 = {"小明" : 190, "小王" : 173, "小美" : 152, 'Amy': 155}
heights.update(heights_2)
print(heights)
```

#### 「修改」

```python
heights['Sam'] = 190
print(heights)
```

#### 「刪除」
使用del指令，並且於 [ ] 符號中輸入要刪除的元素鍵(Key)名稱。

```python
#方法一：利用pop()將指定的key所對應到的value刪除
heights.pop("小美")
print(heights, len(heights))

#方法二：利用del將指定的key所對應到的value刪除
del heights['Amy']
print(heights, len(heights))
```

### Dict的其他特性

```python
#使用keys()、values()、items()來取得所有的鍵、值、項目
print(heights.keys())
print(heights.values())
print(heights.items())

#查看該資料是否在Dict中，是在Key中尋找，不是Value
print("Mike" in heights)
print("Mike" not in heights)

#透過dict()轉換型態
grades = [["國文", 99], ["英文", 60], ["數學", 85]]
print(dict(grades))
```
